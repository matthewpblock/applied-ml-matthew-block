```python
import pandas as pd
```

```python
file_path = 'coral_reef_sites.csv'

try:
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(file_path)

    # Print the first 5 rows of the DataFrame to verify it's loaded correctly
    print("Successfully loaded coral_reef_sites.csv into a DataFrame.")
    print("Here are the first 5 rows:")
    print(df.head())

except FileNotFoundError:
    print(f"Error: The file could not be found at the path: {file_path}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# Check for missing values in the 'coral_cover' column
missing_coral_cover = df['coral_cover'].isnull().sum()

print(f"Number of missing values in 'coral_cover': {missing_coral_cover}")

# To see the percentage of missing values
total_rows = len(df)
if total_rows > 0:
    percentage_missing = (missing_coral_cover / total_rows) * 100
    print(f"Percentage of missing values in 'coral_cover': {percentage_missing:.2f}%")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a style for the plot for better aesthetics
sns.set_style("whitegrid")

# Create a figure for the histogram
plt.figure(figsize=(10, 6))

# Create the histogram for the 'coral_cover' column
# We'll use seaborn's histplot which is powerful and looks great.
# `kde=True` adds a density curve to show the distribution shape.
sns.histplot(data=df, x='coral_cover', bins=30, kde=True)

# Add a title and labels to make the plot informative
plt.title('Distribution of Coral Cover', fontsize=16)
plt.xlabel('Coral Cover', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Display the plot
plt.show()
```

```python
# Check for missing values in the 'coral_cover' column
missing_water_temperature = df['water_temperature'].isnull().sum()

print(f"Number of missing values in 'water_temperature': {missing_water_temperature}")

# To see the percentage of missing values
total_rows = len(df)
if total_rows > 0:
    percentage_missing = (missing_water_temperature / total_rows) * 100
    print(f"Percentage of missing values in 'water_temperature': {percentage_missing:.2f}%")
```

```python
# Create a figure for the histogram
plt.figure(figsize=(10, 6))

# Create the histogram for the 'coral_cover' column
# We'll use seaborn's histplot which is powerful and looks great.
# `kde=True` adds a density curve to show the distribution shape.
sns.histplot(data=df, x='water_temperature', bins=30, kde=True)

# Add a title and labels to make the plot informative
plt.title('Distribution of Water Temperature', fontsize=16)
plt.xlabel('Water Temperature', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Display the plot
plt.show()
```

```python
# Define the conversion factor for meters to feet
M_TO_FT = 3.28084

# Convert 'depth' from meters to feet and create a new column
df['depth_ft'] = df['depth'] * M_TO_FT

# Convert 'water_temperature' from Celsius to Fahrenheit and create a new column
df['water_temperature_f'] = (df['water_temperature'] * 9 / 5) + 32

# Display the first few rows with original and new columns to verify the transformation
print("DataFrame with new columns for converted units:")
print(df[['depth', 'depth_ft', 'water_temperature', 'water_temperature_f']].head())

# You can also check the summary statistics of the new columns
print("\nSummary statistics for the new columns:")
print(df[['depth_ft', 'water_temperature_f']].describe())
```
